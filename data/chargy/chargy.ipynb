```python
import pandas as pd
import geopandas as gpd
import json
import shutil
```

Following [this post](https://docs.astraea.earth/hc/en-us/articles/360043923831-Read-a-KML-File-into-a-GeoPandas-DataFrame) let's read Chargy's KML file using. This may require calling `read_file` without the driver bit first, and then running the `gpd.io` line:

```python
gpd.io.file.fiona.drvsupport.supported_drivers['KML'] = 'rw'
dta = gpd.read_file('https://data.public.lu/fr/datasets/r/22f9d77a-5138-4b02-b315-15f306b77034', driver='KML')
```

```python
dta.explore()
```

```python
dta['x'] = dta.geometry.x
dta['y'] = dta.geometry.y
```

```python
props = dta['Description'].str.extract('<b>(\d+)</b> connectors with (\d+)kW')
dta['connectors'] = props[0].astype(int)
dta['power'] = props[1].astype(int)
```

```python
d = list(dta.drop(['Description', 'geometry'], axis=1).rename(columns={'Name': 'name'}).to_dict(orient='index').values())
file = '../../public/data/chargy/stationmap.json'
with open(file, 'w', encoding='utf-8') as f:
    json.dump(d, f)
```
